```python
import torch
one_d = torch.arange(2, 7)
```

```python
one_d #One dimensional
```

```python
one_d = torch.arange(2, 7, 2) #second 2 provides step size between numbers
one_d
```

```python
one_d = torch.arange(0, 9)
two_d = one_d.view(3,3)
two_d
```

```python
#Now we have a 2D matrix
```

```python
two_d.dim()
```

```python
two_d[1, 2] #The number 5 is located in the second row (1) and third column (2)
```
